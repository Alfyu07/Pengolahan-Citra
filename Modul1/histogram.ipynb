```
import numpy as np
import matplotlib.pyplot as plt
import cv2
from google.colab.patches import cv_imshow
```

```
img = cv2.imread('Image/lena_gray_256.tif')
cv_imshow(img)
```

```
#membuat histogram citra
l = np.arange(img.max()+1)
nk = np.zeros(img.max()+1)

for i in img:
  for j in i:
    nk[j]+=1
```

```
plt.xlabel('keabuan')
plt.ylabel('intensitas')
plt.bar(l,nk)
plt.show()
```
