```
img = cv2.imread('../Image/lena_gray_256.tif')
gray = cv2.cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

```
blue,green,red = cv2.split(img)
```

```
edge = np.array([[1,0,-1],[0,0,0],[-1,0,1]])
sharpen = np.array([[0,-1,0],[-1,5,-1],[0,-1,0]])
```

```
def konvol3d(array,kernel):
  imgPad = np.pad(array,(1,1))
  result = np.zeros(array.shape)
  row,col = result.shape
  for i in range(row):
    for j in range(col):
      result[i][j] += imgPad[i,j] * kernel[0,0] + imgPad[i,j+1] * kernel[0,1] + imgPad[i,j+2] * kernel[0,2] + imgPad[i+1,j] * kernel[1,0] + imgPad[i+1,j+1] * kernel[1,1] + imgPad[i+1,j+2] * kernel[1,2] + imgPad[i+2,j] * kernel[2,0] + imgPad[i+2,j+1] * kernel[2,1] + imgPad[i+2,j+2] * kernel[2,2]
  return result
```

```
#normalisasi
def normalisasi(array):
  max = array.max()
  min = array.min()
  return np.around((array-min)*255/(max-min))
```

```
#konvolusi
result = konvol3d(gray,edge)
cv_imshow(result)
```

```
#normalisasi
result = normalisasi(result)
cv_imshow(result)
```
